## sel vs gen

```python
file = uproot.open("sel_vs_gen.root")
print(file.keys())
print(file.classnames())
sel_hist = file[file.keys()[0]]
gen_hist = file[file.keys()[1]]


sel_hist_val = sel_hist.values()
gen_hist_val = gen_hist.values()

sel_hist_edges = sel_hist.axis().edges()
gen_hist_edges = gen_hist.axis().edges()

# print(len(sel_hist_val))
print(sel_hist_edges[0])
print(sel_hist_edges[-1])

print(gen_hist_edges[0])
print(gen_hist_edges[-1])
```

```python
f, axs = plt.subplots(1,1, figsize=(12,10))

hep.style.use(hep.style.CMS)
hep.cms.label("",com = 13.6, rlabel="Phase-2 (14 Tev, 200 PU)",  fontsize=25, italic=(0,1))

# hep.histplot(sel_hist_val[15:-15], bins=sel_hist_edges[15:-15], ax=axs, density=True, histtype="fill", label="selected triplet", linewidth=2, edgecolor="blue", color="lightblue")
hep.histplot(gen_hist_val, bins=gen_hist_edges, ax=axs, density=True, histtype="fill", label="generated signal", linewidth=2, edgecolor="red", color="peachpuff")

plt.legend(loc='upper right', frameon=False, title='$m_{3\pi}$ histogram', title_fontsize='20', fontsize='18')
plt.xlabel('$m_{3\pi}$ (GeV/$c^2$)', fontsize='20')
plt.ylabel('Normalized Events/2 GeV/$c^2$', fontsize='20')
plt.tick_params(axis='both', labelsize=18) 

plt.ticklabel_format(axis='y', style='plain', scilimits=(0, 0), useOffset=False, useMathText=True)
# plt.gca().yaxis.get_offset_text().set_position((-0.07,1))
# plt.gca().yaxis.get_offset_text().set_fontsize(18)
# plt.show()


plt.savefig("figures/gen_in_acceptance.pdf", dpi=150)
# plt.savefig("figures/selected_signal.pdf", dpi=150)
# plt.savefig("figures/selected_signal.pdf", dpi=150)

```

# Efficiency

```python
file_name = "../histograms/Efficiency_All.root"
file = uproot.open(file_name)
print(file.keys())
print(file.classnames())
gen_hist = file[file.keys()[0]]
reco_hist = file[file.keys()[1]]

gen_hist_val = gen_hist.values()
reco_hist_val = reco_hist.values()

gen_hist_edges = gen_hist.axis().edges()
reco_hist_edges = reco_hist.axis().edges()
```

```python
f, axs = plt.subplots(1,1, figsize=(12,10))

hep.style.use(hep.style.CMS)
hep.cms.label("",com = 13.6, rlabel="Phase-2 (14 Tev, 200 PU)",  fontsize=25, italic=(0,1))

hep.histplot(gen_hist_val[2:],  bins=gen_hist_edges[2:],   ax=axs, density=False, histtype="fill", label="generated pions", linewidth=1.8, edgecolor="blue", color="lightblue", zorder=0)
hep.histplot(reco_hist_val[2:], bins=reco_hist_edges[2:],  ax=axs, density=False, histtype="fill", label="reconstructed pions", linewidth=2, edgecolor="red", color="peachpuff")

plt.legend(loc='upper right', frameon=False, title='$p_{T}$ histogram', title_fontsize='20', fontsize='18')
plt.xlabel('$p_{T}$ (GeV)', fontsize='20')
plt.ylabel('Events', fontsize='20')
plt.tick_params(axis='both', labelsize=18) 

plt.ticklabel_format(axis='y', style='plain', scilimits=(0, 0), useOffset=False, useMathText=True)
# plt.gca().yaxis.get_offset_text().set_position((-0.07,1))
# plt.gca().yaxis.get_offset_text().set_fontsize(18)
# plt.show()


plt.savefig("figures/efficiency_hist.pdf", dpi=150)

```

```python
hep.style.use(hep.style.CMS)
hep.cms.label("",com = 13.6, rlabel="Phase-2 (14 Tev, 200 PU)",  fontsize=25, italic=(0,1))
plt.ticklabel_format(axis='y', style='plain', scilimits=(0, 0), useOffset=False, useMathText=True)

plt.scatter(ranges, efficiency)
plt.plot(ranges, efficiency)

# plt.legend(loc='upper right', frameon=False, title='$p_{T}$ histogram', title_fontsize='20', fontsize='18')
plt.xlabel('$p_{T}$ intervals(GeV)', fontsize='18')
plt.ylabel('Efficiency', fontsize='20')
plt.tick_params(axis='both', labelsize=16) 
plt.savefig("figures/efficiency_vs_pt.pdf", dpi=150)
```
